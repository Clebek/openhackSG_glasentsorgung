```python
# creating a custom function to inspect on which time segments nan values are present

from datetime import timedelta
def broken_time(df_device_id):
    global start_time, end_time, broken_list
    
    # Reset variables
    start_time = None
    end_time = None
    broken_list = []
    
    for i in df_device_id["date_uct"]:
        if start_time is None:
            # First timestamp
            start_time = i
            broken_list.append(f"Start Date: {start_time}")
            end_time = start_time
            
        # checking if the prior value = the value now    
        elif i == end_time:
            continue

        # checking if the prior value + 1 day = the value now   
        elif i == end_time + timedelta(days=1):
            end_time += timedelta(days=1)
        
        # checking if the prior value + 2 = the value now   
        elif i == end_time + timedelta(days=2):
            end_time += timedelta(days=2)

        else:
            broken_list.append(f"End Date: {end_time}")
        
            start_time = i
            broken_list.append(f"Start Date: {start_time}")
            end_time = start_time

    if end_time is not None:
        # End the last charge if applicable
        broken_list.append(f"End Date: {end_time}")

    return broken_list

broken_time(df96)
```

```python
# combining the broken time_df, the broken_time and the create_dataframe functions into one cell

import pandas as pd
from datetime import timedelta, datetime

def broken_time_df(broken_list):
    data = {'Start Date': [], 'End Date': []}
    current_start_date = None

    for item in broken_list:
        # Split each item into its type (Start or End) and the date
        date_type, date_value = item.split(': ')
        date_value = datetime.strptime(date_value, '%Y-%m-%d').date()

        # Check if it's a Start Date or End Date and update the data dictionary
        if date_type == 'Start Date':
            current_start_date = date_value
        elif date_type == 'End Date':
            data['Start Date'].append(current_start_date)
            data['End Date'].append(date_value)

    # Create a DataFrame from the dictionary
    df = pd.DataFrame(data)
    return df

def broken_time(df_device_id):
    global start_time, end_time, broken_list
    
    # Reset variables
    start_time = None
    end_time = None
    broken_list = []
    
    for i in df_device_id["date_uct"]:
        if start_time is None:
            # First timestamp
            start_time = i
            broken_list.append(f"Start Date: {start_time}")
            end_time = start_time

        elif i == end_time:
            continue

        elif i == end_time + timedelta(days=1):
            end_time += timedelta(days=1)

        elif i == end_time + timedelta(days=2):
            end_time += timedelta(days=2)

        else:
            broken_list.append(f"End Date: {end_time}")
         
            start_time = i
            broken_list.append(f"Start Date: {start_time}")
            end_time = start_time

    if end_time is not None:
        broken_list.append(f"End Date: {end_time}")

    return broken_list

def create_dataframes(main_df, column_name, list_of_values):
    # Create an empty dictionary to store DataFrames
    dfs = {}

    # Iterate through the list and create DataFrames
    for i, value in enumerate(list_of_values):
        # Create a new DataFrame based on the condition
        filtered_df = main_df.loc[main_df[column_name] == value].copy()

        # Apply the broken_time function to the filtered DataFrame
        broken_list = broken_time(filtered_df)

        # Convert the broken_list into a DataFrame using broken_time_df
        broken_df = broken_time_df(broken_list)

        # Add the new DataFrame to the dictionary
        dfs[f'df_{i + 1}'] = broken_df

    return dfs



# Call the function to create 25 DataFrames with broken_time and broken_time_df logic applied
result_dfs = create_dataframes(df_nan, 'device_id', nan_device_id_list)
print(result_dfs)

```
