```python
# calculation of nan % for all ids

df = pd.read_csv("./fuellstandsensoren-glassammelstellen-gruenglas.csv", sep = ";")

# Create a list of device_ids
device_ids = nan_device_id_list  # Replace with your actual device_ids


# Step 3: Loop through each device_id
for device_id in device_ids:
    # Filter the DataFrame for the current device_id
    device_df = df.loc[df['device_id'] == device_id]

    # Calculate the number of rows before dropping NaN values
    rows_before = len(device_df)

    # Calculate the number of rows after dropping NaN values
    rows_after = len(device_df.dropna())

    # Step 4: Calculate the percentage of NaN values
    percentage_nan = ((rows_before - rows_after) / rows_before) * 100

    # Print the results
    print(f"For device_id {device_id}:")
    print(f"  Rows before dropping NaN values: {rows_before}")
    print(f"  Rows after dropping NaN values: {rows_after}")
    print(f"  Percentage of NaN values: {percentage_nan:.2f}%")
    print()
```
